```python
import numpy as np
```

```python
n = 3
to_degree = 180 / np.pi

gap_theta = 2 * np.pi / n
```

```python
gap_theta
```

```python
gap_theta * to_degree  # Convert to degrees
```

```python
theta = [gap_theta * n for n in range(n)] # range(3) => [0, 1, 2]
theta
```

```python
[i * to_degree for i in theta]  # Convert each angle to degrees
```

```python
r = 3
x = [r * np.cos(theta_i) for theta_i in theta]
y = [r * np.sin(theta_i) for theta_i in theta]

import matplotlib.pyplot as plt

plt.scatter(x, y)
plt.axis('equal')
```

```python
def calc_position(n, r):
    to_degree = 180 / np.pi
    gap_theta = 2 * np.pi / n
    theta = [gap_theta * i for i in range(n)]
    x = [r * np.cos(theta_i) for theta_i in theta]
    y = [r * np.sin(theta_i) for theta_i in theta]
    return x, y

def draw_pos(x, y):
    plt.scatter(x, y)
    plt.axis('equal')
    plt.show()

X, Y = calc_position(30, 3)
draw_pos(X, Y)
```
